### 8.1 文件的读写

### 8.1.1 文件的打开
    文件打开的通用格式

```python
with open("文件路径"，"打开模式",encoding = "文件操作符编码") as f:
    """对文件进行操作，使用with代码段结束后自动关闭文件"""
```

#### 文件路径
    使用完整绝对路径
    如果是同级文件夹可以省略，直接使用文件名称

```python
with open("test.txt","r", encoding = "utf-8") as f:
    txt = f.read()
    print(txt, type(txt))
    
```

#### 打开模式
    "r" 只读模式，如文件不存在会报错 
    "w" 覆盖写模式，如文件不存在会创建，如果存在会完全覆盖
    "x" 创建写模式，如文件不存在会创建，如果存在会报错
    "a" 追加写模式，如文件不存在会创建，如果存在会在后面追加写
    "b" 二进制文件模式，不能单独使用需要配合 rb wb ab 该模式不需要指定 encoding
    "t" 文本文件模式，是默认值， rt wb ab
    "+" 在原功能上追加

#### 打开模式缺省则默认为 rt

#### 字符编码 Windows环境下默认为gbk
    尽量声明清楚编码方式

### 8.1.2 文件读取

#### 1. 读取整个内容—— f.read()

```python
with open("test.txt","r", encoding = "utf-8") as f:
    txt = f.read()
    print(txt, type(txt))
    
```

```python
with open("test.txt","r", encoding = "gbk") as f:
    txt = f.read()
    print(txt, type(txt))
```

#### 2. 逐行读取 —— f.readline()

```python
with open("test.txt","r", encoding = "utf-8") as f:
    txt = f.readline()
    print(txt, type(txt))
```

```python
with open("test.txt","r", encoding = "utf-8") as f:
    while True:
        text = f.readline()
        if not text:  # 这样可以判断是否为读取到行尾，读取到空行时不会终止循环因为空行为 \n
            break
        else:
            print(text,end="") # 打印的时候会打印txt中本身的\n 换行符，所以需要取消print产生的换行符
```

#### 3. 读入所有行，以每行为元素形成一个列表 —— f.readlines()

```python
with open("test.txt","r", encoding = "utf-8") as f:
    txt = f.readlines()
    print(txt, type(txt))
    print(len(txt))
```

#### 4. 文本文件描述符 f 本身就是可迭代对象

```python
with open("test.txt","r", encoding = "utf-8") as f:
    for txt in f:
        print(txt)
```

#### 二进制文件

```python
with open("test.jpg","rb") as f:
    jpg = f.readlines()
    print(len(jpg))
    for lines in jpg:
        print(lines,end="")
    
```

### 8.1.3 文件的写入

#### 向文件写入一个自负床或者字节流(二进制) —— f.write()

    "w" 覆盖写模式，如文件不存在会创建，如果存在会完全覆盖

```python
with open("write.txt","w",encoding="utf-8") as f:
    f.write("test123")
    f.write("if next")
    print(f.read())   # 想要读取需要加读权限
```

#### 既读又写
>r+  
>如果文件不存在会报错\
>打开文件时指针在文件开头，如果直接写入会直接从文件开头进行覆盖
    
>w+
如果文件不存在则创建\
打开文件时之前的内容会被直接清空

>a+
如果文件不存在则创建\
指针在末尾，不会清空之前的内容

    F.seek方法
    作用:
    设置读写位置
    F.seek(偏移量, whence=相对位置)
    偏移量
    大于0的数代表向文件末尾方向移动的字节数
    小于0的数代表向文件头方向中移动的字节数
    相对位置
    0 代表从文件头开始偏移
    1 代表从文件当前读写位置开始偏移
    2 代表从文件尾开始偏移

```python
with open("write.txt","w+",encoding="utf-8") as f:
    f.write("test123\n")
    f.write("if next")
    f.seek(0,0)
    print(f.read())
```

### 8.1.5 数据读取与存储

####  格式化存储

1.csv格式存储

    由 "," 进行隔开
    由 "\n" 进行换行
    可以由excel打开
    例：
    姓名,语文,数学,英语
    wang,43,65,74
    tan,44,74,35
    lin,23,75,34

2.json格式

    需要依靠 json 模块
    用于记录字典格式
    写入 —— dump()
    读取 —— load()

```python
import json

scores = {"wang":{"math":43,"eng":64},"tan":{"math":34,"eng":67},"lin":{"math":34,"eng":65}}

with open("jsontest.json","w+",encoding="utf-8") as f:
    json.dump(scores,f,indent=4, ensure_ascii=False)
```

```python
import json

with open("jsontest.json","r",encoding="utf-8") as f:
    scores_2 = json.load(f)
    print(type(scores_2))
    for k,v in scores_2.items():
        print(k,v)
```
